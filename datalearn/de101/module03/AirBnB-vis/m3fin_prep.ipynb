### Listing table preparation

```python
import pandas as pd
import numpy as np
```

Open 'listing' table

```python
df = pd.read_csv('./listings.csv')
```

Select columns what we will need for analyze

```python
result_listing = df[['id', 'listing_url', 'host_id', 'neighbourhood', 'latitude', 'longitude', 'bathrooms', 'bedrooms', 'beds', 'bed_type', 'square_feet',
                     'price', 'weekly_price', 'monthly_price', 'security_deposit', 'cleaning_fee', 'minimum_nights', 'maximum_nights',
                     'availability_365', 'review_scores_rating']]
result_listing
```

We can see some NaN values in prices. we will fill it as 0.0 for future calculations

```python
result_listing = result_listing.fillna(0.0)
```

For calculations we need to convert our prices to 'float'. 

```python
def price_to_float(price):
    if type(price) == str:
        price = price.replace('$', '')
        price = price.replace(',', '')
        return float(price)
    else:
        return price

result_listing['price'] = result_listing['price'].map(price_to_float)
result_listing['weekly_price'] = result_listing['weekly_price'].map(price_to_float)
result_listing['monthly_price'] = result_listing['monthly_price'].map(price_to_float)
result_listing['security_deposit'] = result_listing['security_deposit'].map(price_to_float)
result_listing['cleaning_fee'] = result_listing['cleaning_fee'].map(price_to_float)
```

We will mean what if object has 0 in weekly and monthly prices, it has no discount for long time rent. So, we will calculate it with this way:
night price * 7 as weekly price and
night price * 31 as monthly price

```python
result_listing['weekly_price'] = (result_listing['price'] * 7).where(result_listing['weekly_price'] == 0, result_listing['weekly_price'])
result_listing['monthly_price'] = (result_listing['price'] * 31).where(result_listing['monthly_price'] == 0, result_listing['monthly_price'])
result_listing
```

Save result to our prapared table

```python
result_listing.to_csv('./result_listing.csv', index=False)
```
